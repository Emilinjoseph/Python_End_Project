# 1.How many are there in each Team and the percentage splitting with respect to the total employees.






```python
total_emp=len(df.index)
print(((df["Team"].value_counts()/total_emp)*100).round(2))
```

# 2.Segregate the employees w.r.t different positions.

```python
arr = df["Position"].unique()
arr
for i in arr:
    print("Position :",i)
    print(df[df['Position']==i].Name)
```

# 3.Find from which age group most of the employees belong to.

```python
 print("The most of the employees belong to the age group of ",df.Age.mode().iloc[0])
```

# 4.Find out under which team and position, spending in terms of salary is high.

```python
max=df.Salary.max()
dff=df[df['Salary']==max]
print("The team {} and the position {} spending in terms of salary".format(dff.Team.iloc[0],dff.Position.iloc[0]))
```

# 5.Find if there is any correlation between age and salary , represent it visually.

```python

df1=df.drop(['Name', 'Team', 'Position','College', 'Height'], axis='columns')
# Compute correlation matrix
corr = df1.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")

# Add title
plt.title("Correlation Matrix ")

# Show plot
plt.show()
```
